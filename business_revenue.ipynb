```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Gaelim/youtube/master/revenue_mean.csv"

df = pd.read_csv(url)

df.head()
```

```python
df = df.drop("Unnamed: 2", axis=1)

df.columns = df.columns.str.lower()

df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
from pandasql import sqldf


def make_select(query: str) -> pd.DataFrame | None:
   return sqldf(query)
```

```python
df["month"] = df["date"].dt.strftime('%Y-%m')

monthly_revenue = df.groupby("month")["revenue"].sum().reset_index()

monthly_revenue["growth_rate"] = ((monthly_revenue["revenue"] -  monthly_revenue["revenue"].shift(1)) / monthly_revenue["revenue"].shift(1) * 100).round(2)

monthly_revenue
```
